# Visualización de datos con Python

La visualización es una parte importante del conjunto de habilidades de todo científico. Consiste en mostrar la información de forma agradable y fácil de digerir. En este taller aprenderemos los aspectos básicos de la visualización de datos



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## gráficas basicas de matplotlib
con matplotlib se pueden crear visualizaciones de barras, dispersion, lineas, entre otras. En este apartado se generan algunas con datos artificialmente generados


```python
# uso de graficas basicas de matplotlib
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
```

Si bien se logra apreciar el gráfico, se pueden agregar más detalles, para mostrar correctamente información


```python
plt.plot(x, y)
plt.grid()
plt.title("Función sinusoidal")
plt.xlabel("Tiempo[s]")
plt.ylabel("Amplitud")
```

```python
# graficando los puntos con scatter
plt.scatter(x, y)
plt.grid()
plt.title("Función sinusoidal")
plt.xlabel("Tiempo[s]")
plt.ylabel("Amplitud")
```

```python
# matplotlib crea objetos gráficos a los que se les pueden añadir muchos dibujos traslapados. Por ejemplo dibujar los puntos y las líneas
plt.plot(x, y)
plt.scatter(x, y, c="r", marker="*") # controlar el color del punto y el tipo de marcador.
plt.grid()
plt.title("Función sinusoidal")
plt.xlabel("Tiempo[s]")
plt.ylabel("Amplitud")
```

```python
# En caso de tener varios dibujos se pueden colocar leyendas para facilitar la lectura
y2 = np.cos(x)
plt.plot(x, y)
plt.plot(x, y2)
plt.grid()
plt.title("Función sinusoidal")
plt.xlabel("Tiempo[s]")
plt.ylabel("Amplitud")
plt.legend(["Funcion seno", "Funcion coseno-"])
```

```python
# gráficos de barras
amplitudes = [1, 2, 3, 4, 5, 6, 7, 8]
etiquetas = ["uno", "dos", "tres", "cuatro", "cinco", "seis", "siete", "ocho"]
plt.bar(etiquetas, amplitudes)
plt.grid()

# barras agrupadas
data = pd.DataFrame({'España' : [826, 943, 942, 901],
                     'Colombia': [668, 781, 791, 813],
                     'México': [488, 553, 563, 537]},
                    index=['Lunes', 'Martes', 'Miercoles', 'Jueves'])
plt.show()
total = data.sum(axis=1)
plt.bar(total.index, total)
plt.show()

# barras horizontales
plt.barh(total.index, total)
plt.show()

# barras agregadas
plt.bar(data.index, data.España + data.Colombia + data.México, label='España')
plt.bar(data.index, data.Colombia + data.México, label='Colombia')
plt.bar(data.index, data.México, label='México')
plt.legend(loc='best')
plt.show()

# barras adyacentes
n = len(data.index)
x = np.arange(n)
width = 0.25
plt.bar(x - width, data.España, width=width, label='España')
plt.bar(x, data.Colombia, width=width, label='Colombia')
plt.bar(x + width, data.México, width=width, label='México')
plt.xticks(x, data.index)
plt.legend(loc='best')
plt.show()
```

```python
# graficos de torta:
manzanas = [20,10,25,30]
nombres = ["Ana","Juan","Diana","Catalina"]
plt.pie(manzanas, labels=nombres)
plt.show()
# con porcentaje
manzanas = [20,10,25,30]
nombres = ["Ana","Juan","Diana","Catalina"]
plt.pie(manzanas, labels=nombres, autopct="%0.1f %%")
plt.axis("equal")
plt.show()

# con rebanadas extraidas
manzanas = [20,10,25,30]
nombres = ["Ana","Juan","Diana","Catalina"]
colores = ["#EE6055","#60D394","#AAF683","#FFD97D","#FF9B85"]
desfase = (0, 0, 0, 0.1)
plt.pie(manzanas, labels=nombres, autopct="%0.1f %%", colors=colores, explode=desfase)
plt.axis("equal")
plt.show()
```

```python
def sinplot(flip=1):

    x = np.linspace(0, 15, 100)

    for i in range(1, 5):

        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

sns.set_style("white")
sns.set_palette("husl")
sinplot()
plt.show()
```

```python
# graficando una distribución (de total bill)
sns.histplot(tips['total_bill'], kde=True)
plt.show()
```

```python
sns.set_palette("Set2") # cambio de paleta
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.grid()
plt.show()
```

```python
# visualizaciones de las relaciones entre cada par de variables con pairplot
sns.set_palette(sns.color_palette("Set2"))
sns.set_style("ticks")
sns.pairplot(tips)
plt.show()
```

```python
# gráficos para variables categóricas
sns.set_palette("pastel")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.grid()
plt.show()
```

```python
# diagramas radiales (radar chart)
# Conjunto de datos de muestra
df = pd.DataFrame({
'group': ['A','B','C','D'],
'var1': [38, 1.5, 30, 4],
'var2': [29, 10, 9, 34],
'var3': [8, 39, 23, 24],
'var4': [7, 31, 33, 14],
'var5': [28, 15, 32, 14]
})

# saber cuantas columnas hay en el dataset
categories=list(df)[1:]
N = len(categories)

# Graficar la columna A del dataset, hay que extraela
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
values

# Para que quede igualmente separado, se dividen los 360 grados (en radianes) del círculo entre la cantidad de variables
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

# crear el gráfico polar
ax = plt.subplot(111, polar=True)

# Dibujar cada eje por variable
plt.xticks(angles[:-1], categories, color='grey', size=8)

# etiquetas
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
plt.ylim(0,40)

# Graficar los datos
ax.plot(angles, values, linewidth=1, linestyle='solid')

# llenar el área
ax.fill(angles, values, 'b', alpha=0.1)

# Show the g
```

```python
# Gráficos de nube de textos
# Un Gráfico interesante para el análisis de palabras es la nube de textos que permite graficar las palabras de tal forma que el tamaño ayude a visualizar la frecuencia de las palabras.
# Como ejemplo se muestra la fábula de la lechera. En primer lugar, se ven que las palabras más frecuentes son las preposiciones.
# eliminando las preposiciones del texto se puede incluso entender el tema del texto por la frecuencia de las palabras

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of word
text=("Había una vez una niña, hija de un granjero, que ayudaba a sus padres en las tareas de casa y en el cuidado de los animales de la granja. Una mañana, tras recoger la leche de las vacas, la madre de la niña se sintió mal y no se encontraba bien para salir de casa. Entonces, pidió a su hija que llevara la leche al mercado para venderla. La niña, muy responsable, le contestó muy contenta que sí. Y más contenta se quedó cuando su madre le prometió que todo el dinero que ella ganase con la venta de la leche, sería para ella. La niña cogió el cántaro lleno de leche y salió de la granja en dirección al pueblo. Por el camino, ella empezó a hacer planes futuros con lo que ganaría: Cuando yo venda esta leche, compraré trescientos huevos. Los huevos, descartando los que no nazcan, me darán al menos doscientos pollos. Los pollos estarán listos para mercadearlos cuando los precios de ellos estén en lo más alto, de modo que para fin de año tendré suficiente dinero para comprarme el mejor vestido para asistir a las fiestas. Y seguía ensimismada en sus pensamientos: Cuando esté en el baile todos los muchachos me pretenderán, y yo los valoraré uno a uno. Pero en ese momento la niña se despistó y no se dio cuenta de que había una piedra en el medio del camino y acabó tropezando en la piedra y cayendo en el suelo. El cántaro voló por el aire y se rompió derramando toda la leche al suelo. La niña, decepcionada y herida, se levantó y lamentó: - ¡Qué desgracia! Ya no tengo nada que vender, no tendré huevos, ni pollitos, ni vestido... eso me pasa por querer demasiado. Y fue así como la niña, frustrada, se levantó, volvió a la granja y reflexionó sobre la oportunidad que tuvo y que la derramó por el suelo. Moraleja: No seas ambiciosa de mejor y más próspera fortuna, que vivirás ansiosa sin que pueda saciarte cosa alguna. No anheles impaciente el bien futuro, mira que ni el presente está seguro.")

# Create the wordcloud object
text = text.lower()
wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()

# eliminado las preposiciones, artículos, conjunciones, pronombres reflexivos y pronombres posesivos para determinar el tema del texto
palabras = [" a ", " ante ", " bajo ", " cabe ", " con ", " contra ", " de ", " desde ", " durante ", " en ", " entre ", " hacia ", " hasta ", " mediante ", " para ", " por ", " según ", " sin ", " so ", " sobre ", " tras ", " versus ", " vía ", " y ", " el ", " la ", " lo ", " los ", " las ", " que ", " al ", " se ", " un ", " una ", " su ", " le ", " más "]

for palabra in palabras:
    text = text.replace(palabra, " ")


# Create the wordcloud object
wordcloud2 = WordCloud(width=480, height=480, margin=0).generate(text)
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
