Caso de Estudio 1: Visualización de Datos de Pacientes con Enfermedades Crónicas


```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px

# Crear un DataFrame con datos ficticios
data = {
    'Paciente': ['Paciente1', 'Paciente2', 'Paciente3', 'Paciente4', 'Paciente5'],
    'Edad': [35, 42, 55, 30, 48],
    'Género': ['M', 'F', 'M', 'F', 'M'],
    'Diabetes': [1, 0, 1, 0, 1],
    'Hipertensión': [0, 1, 0, 1, 0],
    'VIH': [0, 1, 1, 0, 1],  # Cambio en los datos VIH
    'Nivel Glucosa': [100, 125, 130, 100, 140],
    'Presión Arterial': [120, 130, 140, 110, 135]
}

df = pd.DataFrame(data)

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Definir el layout de la aplicación
app.layout = html.Div([
    html.H1("Visualización de Datos de Pacientes con Enfermedades Crónicas"),
    dcc.Dropdown(
        id='dropdown',
        options=[
            {'label': 'Diabetes', 'value': 'Diabetes'},
            {'label': 'Hipertensión', 'value': 'Hipertensión'},
            {'label': 'VIH', 'value': 'VIH'}
        ],
        value='Diabetes'
    ),
    dcc.Graph(id='graph'),
    dcc.Graph(id='scatter')
])

# Callback para actualizar el gráfico de barras
@app.callback(
    Output('graph', 'figure'),
    [Input('dropdown', 'value')]
)
def update_graph(selected_value):
    fig = px.bar(df, x='Paciente', y=selected_value, text='Paciente', color=selected_value)
    return fig

# Callback para actualizar el gráfico de dispersión
@app.callback(
    Output('scatter', 'figure'),
    [Input('dropdown', 'value')]
)
def update_scatter(selected_value):
    if selected_value == 'Diabetes':
        fig = px.scatter(df, x='Nivel Glucosa', y='Presión Arterial', color='Diabetes', title='Relación entre Nivel de Glucosa y Presión Arterial')
    elif selected_value == 'Hipertensión':  # Cambio en la condición para VIH
        fig = px.scatter(df, x='Nivel Glucosa', y='Presión Arterial', color='Hipertensión', title='Relación entre Nivel de Glucosa y Presión Arterial')
    else:
        fig = px.scatter(df, x='Nivel Glucosa', y='Presión Arterial', color='VIH', title='Relación entre Nivel de Glucosa y Presión Arterial')  # Cambio en el título
    return fig

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```

Caso de Estudio 2: Visualización de Datos de Historiales Educativos de Estudiantes

```python
import pandas as pd
import numpy as np
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go

# Crear un DataFrame con datos ficticios
np.random.seed(0)
num_students = 100
data = {
    'Estudiante': ['Estudiante{}'.format(i) for i in range(1, num_students+1)],
    'Calificación': np.random.randint(60, 100, num_students),
    'Asistencia': np.random.randint(0, 100, num_students),
    'Programa Académico': np.random.choice(['Ingeniería', 'Medicina', 'Administración'], num_students)
}

df = pd.DataFrame(data)

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Definir el layout de la aplicación
app.layout = html.Div([
    html.H1("Visualización de Datos de Historiales Educativos de Estudiantes"),
    dcc.Dropdown(
        id='dropdown',
        options=[
            {'label': 'Calificación', 'value': 'Calificación'},
            {'label': 'Asistencia', 'value': 'Asistencia'}
        ],
        value='Calificación'
    ),
    dcc.Graph(id='graph')
])

# Callback para actualizar el gráfico
@app.callback(
    Output('graph', 'figure'),
    [Input('dropdown', 'value')]
)
def update_graph(selected_value):
    if selected_value == 'Calificación':
        data = [
            go.Box(y=df['Calificación'], name='Calificación'),
        ]
        layout = go.Layout(title='Distribución de Calificaciones', yaxis=dict(title='Calificación'))
    else:
        data = [
            go.Histogram(x=df['Asistencia'], nbinsx=20, name='Asistencia'),
        ]
        layout = go.Layout(title='Distribución de Asistencia', xaxis=dict(title='Asistencia'))

    return {'data': data, 'layout': layout}

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```

Caso de Estudio 3: Visualización de Datos de Transacciones Financieras de Clientes Bancarios

```python
import pandas as pd
import numpy as np
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

# Crear un DataFrame con datos ficticios de transacciones financieras
np.random.seed(42)
num_transactions = 50
clientes = ['Cliente{}'.format(i) for i in range(1, 11)]
data = {
    'Cliente Origen': np.random.choice(clientes, num_transactions),
    'Cliente Destino': np.random.choice(clientes, num_transactions),
    'Monto': np.random.randint(100, 10000, num_transactions)
}

df = pd.DataFrame(data)

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Definir el layout de la aplicación
app.layout = html.Div([
    html.H1("Visualización de Datos de Transacciones Financieras de Clientes Bancarios"),
    dcc.Dropdown(
        id='dropdown',
        options=[{'label': client, 'value': client} for client in clientes],
        value=clientes[0]
    ),
    dcc.Graph(id='graph')
])

# Callback para actualizar el gráfico de barras
@app.callback(
    Output('graph', 'figure'),
    [Input('dropdown', 'value')]
)
def update_graph(selected_client):
    # Filtrar el DataFrame para obtener las transacciones relacionadas con el cliente seleccionado
    filtered_df = df[(df['Cliente Origen'] == selected_client) | (df['Cliente Destino'] == selected_client)]

    # Agrupar por cliente y sumar los montos
    transactions = filtered_df.groupby(['Cliente Origen', 'Cliente Destino']).agg({'Monto': 'sum'}).reset_index()

    # Crear el gráfico de barras horizontales
    fig = go.Figure()
    fig.add_trace(go.Bar(
        y=transactions['Cliente Destino'],
        x=transactions['Monto'],
        orientation='h',
        marker=dict(color='skyblue'),
        name='Transacciones'
    ))

    fig.update_layout(title='Transacciones Financieras de {}'.format(selected_client), yaxis_title='Cliente Destino', xaxis_title='Monto')

    return fig

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```
